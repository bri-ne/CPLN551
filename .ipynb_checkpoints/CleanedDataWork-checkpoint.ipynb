```python
import pandas as pd
import geopandas as gpd
```

```python
BGE = pd.read_csv(r'.\BGE.csv')
```

```python
BGE.head()
```

```python
BGE.sort_values(by= "Avg. Offs", ascending=False)
```

```python
BGE_geom = pd.read_csv(r'.\data\bge_with_geom.csv')
```

```python
BGE_geom.head()
```

```python
BGEgeom = gpd.GeoDataFrame(
    BGE_geom, geometry=gpd.points_from_xy(BGE_geom.stop_lon, BGE_geom.stop_lat)).set_crs('epsg:4326')
```

```python
BGEgeom.crs
```

```python
BGEsubset = BGEgeom[["stop_name","Avg. Ons", "Avg. Offs", "Load", "geometry"]]
```

```python
BGEsubset.crs
```

```python
JustGeoms = BGEgeom.drop_duplicates(subset="stop_name", keep='first')
```

```python
JustGeoms = JustGeoms[["stop_name", "stop_lon", "stop_lat", "geometry"]]
```

```python
JustGeoms
```

```python
JustGeoms = gpd.geodataframe(JustGeoms, geometry=gpd.points_from_xy(JustGeoms.stop_lon, JustGeoms.stop_lat))#.set_crs('epsg:4326')
```

```python
BGEsubsetavg = BGEsubset.groupby(by= "stop_name", as_index=False).mean()
BGEsubsetsum = BGEsubset.groupby(by= "stop_name", as_index=False).sum()
```

```python
BGEsubsetavg.head()
```

```python
BGEsubsetsum.head()
```

```python
BGEsubsetsum.geom = BGEsubsetsum.merge(JustGeoms, on="stop_name")
```

```python
BGEsubsetsum.geom.crs
```

```python
import contextily as cx
```

```python
ax = BGEsubsetsum.geom.plot( 
                     column="Avg. Offs",
                     figsize=(12, 12),
                     legend=True,
                     alpha=1,
                     edgecolor='k',
                     markersize = "Avg. Offs",
                     cmap='BuPu',
                     scheme='QUANTILES',
                     legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)})

cx.add_basemap(ax, source=cx.providers.CartoDB.Voyager, crs=BGEsubset.crs)
ax.set_axis_off()
ax.set(title="Average Offs BGE Buses")
```
